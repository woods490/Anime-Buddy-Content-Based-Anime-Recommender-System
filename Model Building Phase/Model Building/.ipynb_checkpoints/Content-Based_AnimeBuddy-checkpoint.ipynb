## Feature Selection

**For this recommendation system, it needs a feature matrix. This feature matrix will be used to cluster similar animes**

The steps will be:

1. Get the needed features from the main dataframe (anime_id, Genres, Studios, Rating, Type, Producers, Synopsis) and get the needed features from the second dataframe (anime_id and theme)
2. Merged both dataframe based on anime_id to the second dataframe
3. Change the dataframe into matrix for clustering purposes

#### 1. Get the needed features from the main dataframe (anime_id, Genres, Studios, Rating, Type, Producers, Synopsis) and get the needed features from the second dataframe (anime_id and theme)

The reason why only these features are selected is because it makes sense to cluster based on them. For example, if you use the score attributes, low score anime will only be recommended to other low score anime, of course I don't want that to happen. This is the same for the other features that are not selected, like the type, status, duration, etc

```python
main_mini = main_df[['anime_id', 'Genres', 'Rating', 'Type', 'Synopsis', 'Producers', 'Studios']]
secondary_mini = secondary_df[['anime_id', 'themes']]
```

#### 2. Merged both dataframe based on anime_id to the second dataframe

```python
merged_df = pd.merge(main_mini, secondary_mini, on='anime_id', how='inner')

merged_df
```

#### 3. Change the dataframe into matrix for clustering purposes

After looking at the dataframe, there are few steps to make it into features matrix;

1. Concat themes and genres columns then change it into one hot encoded
2. Change producers column into dummies
3. Change studios column into dummies
4. Change source column into dummies
5. Change rating column into dummies
6. Change synopsis into latent semantic features using TruncatedSVD

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import ast

# Task 1: Concat themes and genres columns then change it into one hot encoded
# Function to convert string representation of list to comma-separated string
def convert_to_comma_string(themes_string):
    try:
        themes_list = ast.literal_eval(themes_string)
        return ', '.join(map(str, themes_list))
    except (SyntaxError, ValueError):
        return ''  # Handle cases where the string cannot be evaluated as a list

# Apply the function to the 'themes' column
merged_df['themes'] = merged_df['themes'].apply(convert_to_comma_string)

# Concatenate strings from Column1 and Column2, handling empty strings and strings with only spaces
merged_df['themes_genres'] = merged_df.apply(
    lambda row: (
        row['themes'].strip() + ', ' + row['Genres'].strip()
        if (row['themes'].strip() and row['Genres'].strip())
        else row['themes'].strip() + row['Genres'].strip()
    ),
    axis=1
)

# Split the comma-separated genres/themes into lists
merged_df['themes_genres'] = merged_df['themes_genres'].apply(lambda x: x.split(', ') if x else [])

# Use pd.get_dummies to obtain binary matrix
themes_genres_encoded = pd.get_dummies(merged_df['themes_genres'].explode(), prefix='', prefix_sep='')

# Convert boolean values to integers (0 and 1)
themes_genres_encoded = themes_genres_encoded.astype(int)

# Aggregate the results by grouping on the index (level=0) and summing
themes_genres_encoded = themes_genres_encoded.groupby(level=0).max()

# Task 2: Change producers column into one hot encoded
merged_df['Producers'] = merged_df['Producers'].apply(lambda x: x.split(', ') if x else [])
producers_encoded = pd.get_dummies(merged_df['Producers'].explode(), prefix='', prefix_sep='')
producers_encoded = producers_encoded.astype(int)
producers_encoded = producers_encoded.groupby(level=0).max()

# Task 3: Change studios column into one hot encoded
merged_df['Studios'] = merged_df['Studios'].apply(lambda x: x.split(', ') if x else [])
studios_encoded = pd.get_dummies(merged_df['Studios'].explode(), prefix='', prefix_sep='')
studios_encoded = studios_encoded.astype(int)
studios_encoded = studios_encoded.groupby(level=0).max()

# Task 4: Change type column into one hot encoded
type_encoded = pd.get_dummies(merged_df['Type'].explode(), prefix='', prefix_sep='')
type_encoded = type_encoded.astype(int)
type_encoded = type_encoded.groupby(level=0).max()

# Task 5: Change rating column into one hot encoded
rating_encoded = pd.get_dummies(merged_df['Rating'].explode(), prefix='', prefix_sep='')
rating_encoded = rating_encoded.astype(int)
rating_encoded = rating_encoded.groupby(level=0).max()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import Normalizer
from sklearn.pipeline import make_pipeline
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import matplotlib.pyplot as plt
import re
```

```python
def preprocess_text(text):
    # Remove non-ASCII characters and numbers
    text = ''.join([char for char in text if char.isascii() or char.isspace() or char.isalpha()])
    
    # Remove characters from non-English alphabets
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    
    # Tokenize and lemmatize
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(token.lower()) for token in tokens if token.isalpha()]
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    
    return ' '.join(tokens)
```

```python
# Apply preprocessing to the 'synopsis' column
merged_df['Synopsis'] = merged_df['Synopsis'].apply(preprocess_text)

merged_df
```

```python
# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(merged_df['Synopsis'])

# LSA on TF-IDF Matrix and reduce the dimension into 5% of its original size
svd = TruncatedSVD(n_components=round(X.shape[1]*0.05))
normalizer = Normalizer(copy=False)
lsa = make_pipeline(svd, normalizer)
X_lsa = lsa.fit_transform(X)

# Calculate explained variance
explained_variance = svd.explained_variance_ratio_.sum()

print(f"Reduction to 5% of original size results in: {explained_variance} explained variance of original data" )
```

```python
#Change into dataframe so it can be concacenated
lsa_columns = [f"lsa_component_{i+1}" for i in range(X_lsa.shape[1])]
X_lsa_df = pd.DataFrame(X_lsa, columns=lsa_columns)

X_lsa_df
```

```python
# Concatenate all the encoded DataFrames
feature_matrix = pd.concat([themes_genres_encoded, producers_encoded, studios_encoded, 
                            type_encoded, rating_encoded, X_lsa_df], axis=1)

# Display the resulting feature matrix
print("Feature Matrix:")
print(feature_matrix)
```

## Creating the Recommendation Function

Cluster movies based on feature matrix using PCA-GAKMEANS --> Get other anime that are in the same cluster as queried anime and sort them based on the highest rank

note: for testing purposes get top 10 recommendations

### Clustering using PCA-GAKMEANS

Dimension Reduction Phase

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
```

```python
# Apply PCA
pca = PCA()
pca.fit(feature_matrix)

# Calculate explained variance ratio with percentages
explained_variance_ratio_percentage = np.cumsum(pca.explained_variance_ratio_) * 100

# Set the desired threshold for explained variance
# High Accuracy (Focus on retaining most information) Approach = 95
# Balanced Approach = 85
# Speed (Focus on dimension reduction) Approach = 70

threshold = 70

# Find the number of components needed to exceed the threshold
n_components = np.argmax(explained_variance_ratio_percentage >= threshold) + 1

# Display the result
print(f"Number of components to explain {threshold}% of variance: {n_components}")

# Optionally, you can plot the explained variance ratio
fig, ax = plt.subplots()
ax.plot(range(1, len(explained_variance_ratio_percentage) + 1), explained_variance_ratio_percentage, marker='o')

plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance (%)')
plt.title('PCA - Explained Variance Ratio')
plt.show()
```

Reduce to number of components from the original matrix to represents at least at threshold level of information of the original data

```python
# Apply PCA
pca = PCA(n_components=n_components)
features_transformed = pca.fit_transform(feature_matrix)

features_transformed
```

Do the GAKMEANS with 500 max-k and an 50 increment of K with patience 3

```python
import numpy as np
from sklearn.cluster import KMeans
import random

def initialize_population(population_size, n_clusters, genres_transformed):
    """
    Initialize a population of centroids for K-means clustering.

    Parameters:
    - population_size: Number of individuals in the population
    - n_clusters: Number of clusters (k)
    - genres_transformed: Input data points

    Returns:
    - population: List of randomly initialized centroids
    """
    population = []
    for _ in range(population_size):
        centroids = np.array([random.choice(genres_transformed) for _ in range(n_clusters)])
        population.append(centroids)
    return population

def calculate_fitness(centroids, genres_transformed):
    """
    Calculate the fitness (inertia) of a set of centroids for K-means clustering.

    Parameters:
    - centroids: Centroids to evaluate
    - genres_transformed: Input data points

    Returns:
    - fitness: Inertia of the K-means clustering with the given centroids
    """
    kmeans = KMeans(n_clusters=len(centroids), init=centroids, n_init=1)
    kmeans.fit(genres_transformed)
    return kmeans.inertia_

def crossover(parent1, parent2):
    """
    Perform crossover between two sets of centroids.

    Parameters:
    - parent1: First parent centroids
    - parent2: Second parent centroids

    Returns:
    - child: New set of centroids resulting from crossover
    """
    crossover_point = random.randint(1, len(parent1) - 1)
    child = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
    return child

def mutate(centroids, mutation_rate):
    """
    Mutate a set of centroids.

    Parameters:
    - centroids: Centroids to mutate
    - mutation_rate: Probability of mutation for each centroid

    Returns:
    - mutated_centroids: Mutated centroids
    """
    mutated_centroids = centroids.copy()
    for i in range(len(mutated_centroids)):
        if random.random() < mutation_rate:
            mutated_centroids[i] = random.choice(mutated_centroids)
    return mutated_centroids

def genetic_algorithm(genres_transformed, max_k, population_size, generations, mutation_rate, patience):
    """
    Perform K-means clustering optimization using Genetic Algorithm to find the best k and centroids.

    Parameters:
    - genres_transformed: Input data points
    - max_k: Maximum number of clusters to search for
    - population_size: Number of individuals in the population
    - generations: Number of generations in the genetic algorithm
    - mutation_rate: Probability of mutation for each centroid
    - patience: Number of consecutive increases in inertia to tolerate before stopping

    Returns:
    - best_k: Best number of clusters (k)
    - best_centroids: Best centroids found by the genetic algorithm
    """
    best_fitness = float('inf')
    best_k = 0
    best_centroids = None
    prev_fitness = 0
    consecutive_increases = 0
    fitness_nodecrease_consecutive = 0
    fitness_difference_before = 0

    for k in range(25, max_k + 1, 25):
        print(f"Running Genetic Algorithm for k={k}")

        population = initialize_population(population_size, k, genres_transformed)

        for generation in range(generations):
            fitness_scores = [calculate_fitness(centroids, genres_transformed) for centroids in population]
            parents_indices = np.argsort(fitness_scores)[:2]
            parent1, parent2 = population[parents_indices[0]], population[parents_indices[1]]
            child = crossover(parent1, parent2)
            child = mutate(child, mutation_rate)
            least_fit_index = np.argmax(fitness_scores)
            population[least_fit_index] = child

        best_centroids_index = np.argmin(fitness_scores)
        best_fitness_for_k = fitness_scores[best_centroids_index]
        
        fitness_difference_now = prev_fitness - best_fitness_for_k
        
        if fitness_difference_now > fitness_difference_before:
            fitness_nodecrease_consecutive = 0
        else:
            fitness_nodecrease_consecutive += 1

        if best_fitness_for_k < best_fitness:
            best_fitness = best_fitness_for_k
            best_k = k
            best_centroids = population[best_centroids_index]
            consecutive_increases = 0
        else:
            consecutive_increases += 1

        print(f"Best Fitness for k={k}: {best_fitness_for_k}")
        

        # Check if the decrease in inertia doesn't get better after patience
        if consecutive_increases >= patience or fitness_nodecrease_consecutive >= patience:
            print(f"No significant decrease in inertia for {patience} consecutive times. Stopping the algorithm.")
            break

        prev_fitness = best_fitness_for_k
        
        fitness_difference_before = fitness_difference_now
        
    return best_k, best_centroids

def cluster_data(data, best_k, best_centroids):
    """
    Perform K-means clustering on the data using the best K and centroids.

    Parameters:
    - data: Input data points
    - best_k: Best number of clusters (K)
    - best_centroids: Best centroids found by the genetic algorithm

    Returns:
    - cluster_assignments: Cluster assignments for each data point
    - final_centroids: Final centroids after clustering
    """
    kmeans = KMeans(n_clusters=best_k, init=best_centroids, n_init=1)
    cluster_assignments = kmeans.fit_predict(data)
    final_centroids = kmeans.cluster_centers_
    return cluster_assignments, final_centroids
```

```python
# Set parameters
max_k = 500
population_size = 10
generations = 50
mutation_rate = 0.1
patience = 3  # Set your desired patience value

# Run the genetic algorithm
best_k, best_centroids = genetic_algorithm(features_transformed, max_k, population_size, generations, mutation_rate, patience)

# Cluster the data using the best K and centroids
cluster_assignments, final_centroids = cluster_data(features_transformed, best_k, best_centroids)

print("Best Number of Clusters (k):", best_k)
print("Best Centroids:", best_centroids)
print("Final Cluster Assignments:", cluster_assignments)
print("Final Centroids:", final_centroids)
```

```python
merged_df['cluster'] = cluster_assignments

sorting_df = merged_df[['anime_id', 'cluster']]

sorting_df = pd.merge(sorting_df, main_df, on='anime_id', how='inner')
```

```python
sorting_df = sorting_df[sorting_df['Score'] != 'UNKNOWN'][['Name', 'Premiered', 'Episodes', 'Rating',
                                                          'Genres', 'Synopsis', 'Studios', 'cluster', 'Score',
                                                          'Scored By']]

sorting_df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
sorting_df['Score'] = scaler.fit_transform(sorting_df[['Score']])
sorting_df['Scored By'] = scaler.fit_transform(sorting_df[['Scored By']])

sorting_df['Weighted Score'] = (sorting_df['Score'] * 0.7) + (sorting_df['Scored By'] * 0.3)

sorting_df
```

```python
def get_similar_anime(df, anime_name):
    # Step 1: Query the DataFrame to get the cluster of the specified anime by name
    query_result = df[df['Name'] == anime_name]
    if query_result.empty:
        print(f"Anime with name '{anime_name}' not found.")
        return

    anime_cluster = query_result['cluster'].values[0]

    # Step 2: Filter the DataFrame based on the obtained cluster
    similar_anime = df[df['cluster'] == anime_cluster]

    # Step 3: Sort the filtered DataFrame based on the weighted score
    similar_anime = similar_anime.sort_values(by='Weighted Score', ascending=False)

    # Step 4: Display the top 50 anime in the same cluster
    similar_anime = similar_anime[similar_anime['Name'] != anime_name]
    top_10_anime = similar_anime.head(25)
    print(f"Top 10 Anime in the Same Cluster as '{anime_name}':")
    print(top_10_anime[['Name', 'Premiered', 'Episodes', 'Rating', 'Genres', 'Synopsis', 'Studios']])
    
anime_name_to_query = 'Psycho-Pass'
get_similar_anime(sorting_df, anime_name_to_query)
```
